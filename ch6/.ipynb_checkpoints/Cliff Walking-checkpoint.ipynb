# QLearning Control

```python
class QLControl:
    def __init__(self, nS, nA, terminal=list(range(37,48)), eps=None, initializer=None):
        self.Q = np.random.normal(0.5,0.25, (nS, nA) )

        if initializer!=None:
            self.Q = initializer(nS, nA)
            
        
        for i in terminal:
            self.Q[i] = np.array([0, 0, 0, 0])
        
        self.pi = np.random.randint(0, 4, (16, ), dtype=np.int8)
        self.alpha = 0.2
        self.gamma = 0.9
        self.epsilon = eps
    
    def policy(self, state, eps=None):
        
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        
        
        if np.random.random() >= eps:
            return np.argmax(self.Q[state])
        else:
            return np.random.randint(0, 4)
    def step(self, s, env, step_size, eps=None):
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        a = self.policy(s, eps=eps)
        s_, r, done, reward = env.step(a)
        
        self.Q[s,a] = self.Q[s,a] + step_size*(r + np.max(self.Q[s_]) - self.Q[s, a]   )
        
        return s_, r, done, reward
```

## Cliff walking Convergence

```python
def train(n_episodes, step_size, control, env, verbose=False, verbose_status_check=1):
    Rs = []
    for episode in range(n_episodes):
        if verbose and episode%verbose_status_check==1:
            print(f"\repisode: {episode}")
        s = env.reset()
        done = False
        r_i = 0
        while not done:

            
            s_, r, done, reward = control.step(s,env, step_size)
            
            s = s_
            r_i+=r
        Rs.append(r_i)
    return Rs
```

```python
ql_control = QLControl(env.nS, env.nA, eps=0.0)
```

```python
rs_ql = train(500, 0.5, ql_control, env)
```

```python
plt.figure(figsize=(15,10))
plt.plot(rs_ql)
```

# Sarsa Control

```python
class SarsaControl:
    def __init__(self, nS, nA, terminal=list(range(37,48)), eps=None, initializer=None):
        self.Q = np.random.normal(0.5,0.25, (nS, nA) )
        if initializer!=None:
            self.Q = initializer(nS, nA)
        
        for i in terminal:
            self.Q[i] = np.array([0, 0, 0, 0])
        
        self.pi = np.random.randint(0, 4, (16, ), dtype=np.int8)
        self.alpha = 0.2
        self.gamma = 0.9
        self.epsilon = eps
    
    def policy(self, state, eps=None):
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        if np.random.random() >= eps:
            return np.argmax(self.Q[state])
        else:
            return np.random.randint(0, 4)
    def step(self, s, env, step_size, eps=None):
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        a = self.policy(s, eps=eps)
        s_, r, done, reward = env.step(a)
        
        
#         take second action
        a_ = self.policy(s_, eps=eps)
        
        self.Q[s,a] = self.Q[s,a] + step_size*(r + self.Q[s_, a_] - self.Q[s, a]   )
        
        return s_, r, done, reward
```

```python
def get_moving_average(arr, window_size):
    arr = arr[:]
    for i in range(len(arr)-window_size):
        arr[i] = sum(arr[i:i+window_size])
    return arr
```

```python
def initializer(a, b):
    return np.zeros((a, b))*0.6
sarsa_control = SarsaControl(env.nS, env.nA, eps = 0.0, initializer = initializer)
ql_control = QLControl(env.nS, env.nA, eps = 0.0,initializer = initializer)
rs_sarsa = train(500, 0.5, sarsa_control, env)
rs_ql = train(500, 0.5, ql_control, env)
```

# Expected Sarsa

```python
class ExpectedSarsa:
    def __init__(self, nS, nA, terminal=list(range(37,48)), eps=None, initializer=None):
        self.nS = nS
        self.nA = nA
        self.Q = np.random.normal(0.5,0.25, (nS, nA) )
        if initializer!=None:
            self.Q = initializer(nS, nA)
        
        for i in terminal:
            self.Q[i] = np.array([0, 0, 0, 0])
        
#         self.pi = np.random.randint(0, 4, (16, ), dtype=np.int8)
        self.alpha = 0.2
        self.gamma = 0.9
        self.epsilon = eps
    
    def policy(self, state, eps=None):
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        if np.random.random() >= eps:
            return np.argmax(self.Q[state])
        else:
            return np.random.randint(0, 4)
    
    def pi(self,S, Q):
#         return probabilities for each action
        eps = self.epsilon
        if self.epsilon == None:
            eps = 0.2
        
        probs = np.ones(self.nA)*(eps/self.nA)
        probs[np.argmax(Q[S])] +=1-eps
        return probs
    def step(self, s, env, step_size, eps=None):
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        a = self.policy(s, eps=eps)
        s_, r, done, reward = env.step(a)
        
        
#         take second action
        a_ = self.policy(s_, eps=eps)
        
        self.Q[s,a] = self.Q[s,a] + step_size*(r + sum(self.Q[s_]*self.pi(s_, self.Q)) - self.Q[s, a]   )
        
        return s_, r, done, reward
```

```python
def initializer(a, b):
    return np.zeros((a, b))*0.6

e_sarsa_control = ExpectedSarsa(env.nS, env.nA, eps = 0.2, initializer = initializer)

rs_e_sarsa = train(500, 0.5, e_sarsa_control, env)

plot_graph(get_moving_average(rs_e_sarsa,15), "sarsa")
```

# Double Q Learning

```python
class DoubleQLearning:
    def __init__(self, nS, nA, terminal=list(range(37,48)), eps=None, initializer=None):
        self.nS = nS
        self.nA = nA
        self.Q1 = np.random.normal(0.5,0.25, (nS, nA) )
        self.Q2 = np.random.normal(0.5,0.25, (nS, nA) )
        if initializer!=None:
            self.Q1 = initializer(nS, nA)
            self.Q2 = initializer(nS, nA)
        for i in terminal:
            self.Q1[i] = np.array([0, 0, 0, 0])
            self.Q2[i] = np.array([0, 0, 0, 0])
        
#         self.pi = np.random.randint(0, 4, (16, ), dtype=np.int8)
        self.alpha = 0.2
        self.gamma = 0.9
        self.epsilon = eps
    
    def policy(self, state, Q,eps=None):
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        if np.random.random() >= eps:
            return np.argmax(Q[state])
        else:
            return np.random.randint(0, 4)
    
    def pi(self,S, Q):
#         return probabilities for each action
        eps = self.epsilon
        if self.epsilon == None:
            eps = 0.2
        
        probs = np.ones(self.nA)*(eps/self.nA)
        probs[np.argmax(Q[S])] +=1-eps
        return probs
    def step(self, s, env, step_size, eps=None):
        if eps == None:
            if self.epsilon != None:
                eps = self.epsilon
            else:
                eps = 0.2
        
#         taken action
        a = self.policy(s,self.Q1+self.Q2 ,eps=eps)
#         OBSERVE ENV
        s_, r, done, reward = env.step(a)
        
#         Alternate bw Q Values with equal probability
        if np.random.rand() < 0.5:
            self.Q1 = self.Q1 + step_size*(r + self.Q2[s_, np.argmax(self.Q1[s_])] -self.Q1[s,a])
        else:
            self.Q2 = self.Q2 + step_size*(r + self.Q1[s_, np.argmax(self.Q2[s_])] -self.Q2[s,a])
        
            
#         take second action
        
        return s_, r, done, reward
```

```python
def initializer(a, b):
    return np.zeros((a, b))*0.6
dql_control = DoubleQLearning(env.nS, env.nA, eps = 0.2, initializer = initializer)

rs_dql = train(500, 0.5, dql_control, env)


plot_graph(get_moving_average(rs_dql,15), "Double Q Learning")
```
